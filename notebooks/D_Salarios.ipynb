```python
import pandas as pd

url="https://raw.githubusercontent.com/MarioV3232/etl-data-pipeline2507452022/refs/heads/main/data/raw/D_salarios.csv"
df = pd.read_csv(url)

df.head()
```

```python
# Limpiar Datos
df = df.drop_duplicates()
df = df.dropna()
```

```python
# Separar curated y rejects
df_rejects = df[df.isnull().any(axis=1)]
df_curated = df.dropna()
print(f"Curated: {len(df_curated)} | Rejects: {len(df_rejects)}")
```
